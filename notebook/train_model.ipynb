```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Initialize the Count Vectorizer 
count_vectorizer = TfidfVectorizer(
        stop_words = None,  # Stopwords not Removed
        lowercase = True,   # Text Convert into Lower Case
        analyzer = 'word',    # Word n-grams Generation
        token_pattern = r'\w{2,}', #vectorize 2-character words or more
        ngram_range = (1,1),  # Word Uni-grams Features
        max_features = None)  # Extract All Features 
Corpus = data['SMS'] 


# Fit the Count Vectorizer on Input Text
count_vectorizer.fit(Corpus)

# Transform the Input Text using Count Vectorizer
transform_features = count_vectorizer.transform(Corpus)

# Get the name of Features (Feature  Set)
feature_set = count_vectorizer.get_feature_names_out()

# Convert Transformed features into Array and Create a Dataframe
Input_features = pd.DataFrame(transform_features.toarray(), columns = feature_set)

# Combine the Input Features and Output Label Gender
extracted_features=Input_features.join(output_labels_df)

#Save the Features in CSV File 
#extracted_features.to_csv(r'sample-data-unigram-features.csv', index = False, header=True)

# Display the Extracted Features 
print("\nFeatures Extracted from Text :")
print("==============================\n")
pd.set_option('display.max_columns', None)
extracted_features.head()
```

```python
input_data = extracted_features.iloc[:,:-12]
output_data= extracted_features.iloc[:,-12:]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000,
                           n_informative=2, n_redundant=0,
                           random_state=0, shuffle=True)
clf = RandomForestClassifier(random_state=0)
clf.fit(input_data, output_data)
```
